# Cleaning du dataset

```python
# Pivot est la colonne qui indique le chanteur principal, ici on garde que les ligne où pivot == 0 (chanteur principal)
spotify = spotify[spotify['pivot']==0]
```

```python
# Conversion des colonnes au bon type
columns_to_convert_float = ['valence', 'speechiness','acousticness','loudness','tempo','energy','danceability']
columns_to_convert_int = ['mode', 'rank', 'pivot','streams', 'peak_rank', 'weeks_on_chart', 'key', 'duration']
spotify[columns_to_convert_float] = spotify[columns_to_convert_float].apply(pd.to_numeric).astype(float)
spotify[columns_to_convert_int] = spotify[columns_to_convert_int].apply(pd.to_numeric).astype(int)
spotify['week'] = pd.to_datetime(spotify['week_x'], format='%Y-%m-%d')
```

```python
# Enlever les null et les lignes où la région n'est pas indiqué

spotify = spotify.dropna()
spotify = spotify[spotify['region'] != 'region']
spotify.shape
```

# TOP50

```python
# j'ai enlevé les colonnes qui font que les lignes ne sont pas en doublons
top50 = spotify.drop(columns = ['rank', 'peak_rank', 'weeks_on_chart', 'week' ]) 
```

```python
#sommer les streams par continents
top50 = top50.groupby(['artist_names', 'artist_individual', 'track_name', 'danceability',
       'energy', 'key', 'mode', 'loudness', 'speechiness', 'acousticness'
      , 'valence', 'tempo', 'duration',
     'continent', 'artist_genrefin']).sum()['streams'].reset_index() 
```

```python
#avoir le max de nombre de streams par region
top50 = top50.sort_values(by=['continent', 'streams'], ascending = False ) 
```

```python
#mettre un rang des streams les plus écoutés par continents
top50['sub_group_rank'] = top50.groupby('continent')['streams'].rank(ascending = False) 
```

```python
# garder les top 20 streams par continents
top50 = top50[(top50['sub_group_rank'] > 0) & (top50['sub_group_rank'] < 51)]
```

```python
top50['top_track'] = 'Top Track' 
```

```python
result3 = round((top50['streams'].sum()/spotify['streams'].sum())*100,2)

print(f'La part du top 50 est de {result3}%')
```

# Prédiction du hit parfait

```python
#merge les 2 dataset 
spot = pd.merge(spotify, top50, on = ['artist_names', 'track_name'], how = 'left')
```
